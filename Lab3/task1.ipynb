```python
import numpy as np
from sklearn.neural_network import MLPClassifier
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
#MLPClassifier automatically handles the output layer based on classes.
mlf = MLPClassifier(
    solver='lbfgs',
    alpha=1e-5,
    hidden_layer_sizes=(5, ), #This creates a single, shallow hidden layer.
    random_state=1
    )
mlf.fit(X, y) #Train the model

clf = MLPClassifier(
    solver='lbfgs',
    alpha=1e-5,
    hidden_layer_sizes=(5, 2), # this creates two distinct hidden layers Layer 1: 5 neurons, Layer 2: 2 neurons
    random_state=1
    )
clf.fit(X, y) #Train the model
# Output results
print("Predictions:", mlf.predict(X))
print("\nWeights (Input to Hidden):\n", mlf.coefs_[0])
print("\nBiases (Hidden Layer):\n", mlf.intercepts_[0])
print("\nWeights (Hidden to Output):\n", mlf.coefs_[1])
print("\nBiases (Output Layer):\n", mlf.intercepts_[1])
print("-"*50)
print("Predictions:", clf.predict(X))
print("\nWeights (Input to Hidden):\n", clf.coefs_[0])
print("\nBiases (Hidden Layer):\n", clf.intercepts_[0])
print("\nWeights (Hidden to Output):\n", clf.coefs_[1])
print("\nBiases (Output Layer):\n", clf.intercepts_[1])
print("-"*50)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

mlp = MLPClassifier(hidden_layer_sizes=(5,), activation='relu', solver='lbfgs', random_state=42)
mlp.fit(X, y)

# Create a mesh grid to plot the decision boundary
xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 200), np.linspace(-0.5, 1.5, 200))
Z = mlp.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(6, 5))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='bwr')
plt.scatter(X[y==0][:,0], X[y==0][:,1], color='blue', label='Class 0 (Output 0)', s=100, edgecolor='k')
plt.scatter(X[y==1][:,0], X[y==1][:,1], color='red', label='Class 1 (Output 1)', s=100, edgecolor='k')
plt.title("Task 01: MLP Decision Boundary for XOR Problem")
plt.xlabel("Input X1")
plt.ylabel("Input X2")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
