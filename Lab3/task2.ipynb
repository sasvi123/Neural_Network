```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# 1. Load the California housing dataset
data = fetch_california_housing()
X = data.data
y = data.target

# 2. Split data into 90% training and 10% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# 3. Normalize features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Create an MLP using MLPRegressor with the specified parameters
mlp = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),
    activation='relu',
    solver='adam',
    max_iter=300,
    batch_size='auto',
    random_state=42  # Ensures reproducible results for your report
)

# 5. Train the model
mlp.fit(X_train_scaled, y_train)

# 6. Evaluate on the 10% test split
y_pred = mlp.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error on Test Set: {mse}")

# 7. Visualize the loss curve
plt.figure(figsize=(7, 4.5))
plt.plot(mlp.loss_curve_, color='darkorange', linewidth=2)  # Fixed the variable name here!
plt.title("Task 02: MLPRegressor Training Loss Curve")
plt.xlabel("Iterations (Epochs)")
plt.ylabel("Loss (MSE)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
