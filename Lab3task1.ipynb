```python
import numpy as np
from sklearn.neural_network import MLPClassifier
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
#MLPClassifier automatically handles the output layer based on classes.
mlf = MLPClassifier(
    solver='lbfgs',
    alpha=1e-5,
    hidden_layer_sizes=(5, ), #This creates a single, shallow hidden layer.
    random_state=1
    )
mlf.fit(X, y) #Train the model

clf = MLPClassifier(
    solver='lbfgs',
    alpha=1e-5,
    hidden_layer_sizes=(5, 2), # this creates two distinct hidden layers Layer 1: 5 neurons, Layer 2: 2 neurons
    random_state=1
    )
clf.fit(X, y) #Train the model
# Output results
print("Predictions:", mlf.predict(X))
print("\nWeights (Input to Hidden):\n", mlf.coefs_[0])
print("\nBiases (Hidden Layer):\n", mlf.intercepts_[0])
print("\nWeights (Hidden to Output):\n", mlf.coefs_[1])
print("\nBiases (Output Layer):\n", mlf.intercepts_[1])
print("-"*50)
print("Predictions:", clf.predict(X))
print("\nWeights (Input to Hidden):\n", clf.coefs_[0])
print("\nBiases (Hidden Layer):\n", clf.intercepts_[0])
print("\nWeights (Hidden to Output):\n", clf.coefs_[1])
print("\nBiases (Output Layer):\n", clf.intercepts_[1])
print("-"*50)
```
